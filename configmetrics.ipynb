```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import numpy as np
import matplotlib.pyplot as plt

# Example: input is a single value, valid if >0.5, recommendation is always 0.7
inputs = np.linspace(0, 1, 100)
is_valid = inputs > 0.5
recommendations = np.full_like(inputs, 0.7)

plt.figure(figsize=(10, 6))

# Fill the valid/invalid regions
plt.axvspan(0, 0.5, color='mistyrose', alpha=0.4, label='Invalid Region')
plt.axvspan(0.5, 1, color='honeydew', alpha=0.4, label='Valid Region')

# Plot validity line
plt.plot(inputs, is_valid, label='Valid (True=1, False=0)', color='blue', lw=2)

# Plot input points
plt.scatter(inputs[~is_valid], np.zeros_like(inputs[~is_valid]), color='red', s=30, label='Invalid SUBCONFIG VALUE', zorder=5)
plt.scatter(inputs[is_valid], np.ones_like(inputs[is_valid]), color='green', s=30, label='Valid SUBCONFIG VALUE', zorder=5)

# Plot recommendations for invalid inputs
for x, y in zip(inputs[~is_valid], recommendations[~is_valid]):
    plt.arrow(x, 0, y-x, 1, head_width=0.02, head_length=0.03, fc='orange', ec='orange', alpha=0.5, length_includes_head=True)
    # Annotate a few for demonstration
    if np.isclose(x, 0.1) or np.isclose(x, 0.4):
        plt.text(x, 0.05, f"{x:.2f}→{y:.2f}", fontsize=8, color='orange', rotation=60)

# Formatting
plt.xlabel('Input Configuration Value', fontsize=12)
plt.ylabel('Validity (1=Valid, 0=Invalid)', fontsize=12)
plt.title('Subconfig Validation and Crypto Recommendation Visualization', fontsize=14)
plt.legend(loc='upper left')
plt.ylim(-0.1, 1.1)
plt.xlim(-0.05, 1.05)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# Example network config inputs (in decimal form, e.g., 0.1% = 0.001)
inputs = np.array([1, 2, 4, 5, 9, 10, 15, 20])
min_val, max_val = 4, 6  # 0.1% to 1%

# Validation
is_valid = (inputs >= min_val) & (inputs <= max_val)

# Recommendation engine: recommend nearest valid boundary if out of range
def recommend_value(x, min_v, max_v):
    if x < min_v:
        return min_v
    elif x > max_v:
        return max_v
    else:
        return x

recommendations = np.array([recommend_value(x, min_val, max_val) if not v else None for x, v in zip(inputs, is_valid)])

# Visualization
plt.figure(figsize=(10, 5))
plt.axhspan(min_val, max_val, color='honeydew', alpha=0.6, label='Valid Range [4, 6]')
plt.scatter(np.arange(len(inputs))[is_valid], inputs[is_valid], color='green', s=80, label='Valid Input')
plt.scatter(np.arange(len(inputs))[~is_valid], inputs[~is_valid], color='red', s=80, label='Invalid Input')

# Draw arrows for recommendations
for idx, (x, v, rec) in enumerate(zip(inputs, is_valid, recommendations)):
    if not v:
        plt.scatter(idx, rec, color='orange', s=150, marker='*', label='Recommended' if idx == list(np.where(~is_valid)[0])[0] else "")
        plt.annotate('', xy=(idx, rec), xytext=(idx, x),
                     arrowprops=dict(facecolor='orange', width=2, headwidth=8, alpha=0.6))

plt.xticks(np.arange(len(inputs)), [f"{x}" for x in inputs])
plt.ylabel('guest dhcp lease time Value')
plt.xlabel('Customer Device Index Id ')
plt.title('Validation and Recommendation for guest dhcp lease time subconfig (4 to 20 hrs)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# Example network config inputs (in decimal form, e.g., 0.1% = 0.001)
inputs = np.array([0.0005, 0.001, 0.002, 0.007, 0.009, 0.010, 0.012, 0.02])
min_val, max_val = 0.001, 0.01  # 0.1% to 1%

# Validation
is_valid = (inputs >= min_val) & (inputs <= max_val)

# Recommendation engine: recommend nearest valid boundary if out of range
def recommend_value(x, min_v, max_v):
    if x < min_v:
        return min_v
    elif x > max_v:
        return max_v
    else:
        return x

recommendations = np.array([recommend_value(x, min_val, max_val) if not v else None for x, v in zip(inputs, is_valid)])

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True, gridspec_kw={'height_ratios': [2, 1]})

# --- Top plot: value scatter and recommendations ---
ax1.axhspan(min_val, max_val, color='honeydew', alpha=0.6, label='Valid Range [0.1%, 1%]')
ax1.scatter(np.arange(len(inputs))[is_valid], inputs[is_valid], color='green', s=80, label='Valid Input')
ax1.scatter(np.arange(len(inputs))[~is_valid], inputs[~is_valid], color='red', s=80, label='Invalid Input')

for idx, (x, v, rec) in enumerate(zip(inputs, is_valid, recommendations)):
    if not v:
        ax1.scatter(idx, rec, color='orange', s=150, marker='*', label='Recommended' if idx == list(np.where(~is_valid)[0])[0] else "")
        ax1.annotate('', xy=(idx, rec), xytext=(idx, x),
                     arrowprops=dict(facecolor='orange', width=2, headwidth=8, alpha=0.6))

ax1.set_ylabel('Network Config Value')
ax1.set_title('Validation and Recommendation for Network Config (0.1% to 1%)')
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.5)

# --- Bottom plot: Boolean validity ---
ax2.plot(np.arange(len(inputs)), is_valid.astype(int), 'bo-', lw=2, label="Validity (True=1, False=0)")
ax2.set_yticks([0, 1])
ax2.set_yticklabels(['False', 'True'])
ax2.set_xlabel('Sample Index (Input Value)')
ax2.set_ylabel('Validity')
ax2.set_xticks(np.arange(len(inputs)))
ax2.set_xticklabels([f"{x:.4f}" for x in inputs])
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# Allowed categories
valid_categories = ["red", "green", "blue", "yellow"]
# Example inputs (some valid, some invalid)
inputs = ["red", "purple", "green", "orange", "blue", "pink", "yellow", "brown"]

# Validation
is_valid = [x in valid_categories for x in inputs]

# Simple Recommendation Engine: Closest valid category by Levenshtein distance
def recommend_category(input_value, valid_categories):
    # For demo, use simple string similarity (could use ML in real systems)
    from difflib import get_close_matches
    matches = get_close_matches(input_value, valid_categories, n=1, cutoff=0.0)
    return matches[0] if matches else None

recommendations = [
    recommend_category(x, valid_categories) if not v else None
    for x, v in zip(inputs, is_valid)
]

# Visualization
plt.figure(figsize=(9, 4))
y_pos = np.arange(len(inputs))
colors = ['green' if v else 'red' for v in is_valid]
plt.barh(y_pos, [1]*len(inputs), color=colors, edgecolor='k', alpha=0.5)
plt.yticks(y_pos, inputs)
plt.xlabel("Validity (1=Valid, 0=Invalid)")
plt.title("Categorical Validation and Recommendations")
for i, (valid, rec) in enumerate(zip(is_valid, recommendations)):
    if not valid:
        plt.text(0.5, i, f"→ Recommend: {rec}", color='orange', va='center', fontsize=10)
plt.xlim(0, 1.2)
plt.tight_layout()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# Example: input is a single value, valid if >0.5, recommendation engine suggests minimum valid + margin
inputs = np.linspace(0, 1, 100)
is_valid = inputs > 0.5

# Recommendation engine: for each invalid x, recommend 0.5 + 0.05 * np.random.rand()
np.random.seed(42)
recommendations = np.where(is_valid, inputs, 0.5 + 0.05 * np.random.rand(np.sum(~is_valid)))

plt.figure(figsize=(12, 7))

# Fill the valid/invalid regions
plt.axvspan(0, 0.5, color='mistyrose', alpha=0.4, label='Invalid Region')
plt.axvspan(0.5, 1, color='honeydew', alpha=0.4, label='Valid Region')

# Plot validity line
plt.plot(inputs, is_valid, label='Valid (True=1, False=0)', color='blue', lw=2)

# Plot input points
plt.scatter(inputs[~is_valid], np.zeros_like(inputs[~is_valid]), color='red', s=40, label='Invalid Inputs', zorder=5)
plt.scatter(inputs[is_valid], np.ones_like(inputs[is_valid]), color='green', s=40, label='Valid Inputs', zorder=5)

# Plot recommended values for invalid inputs
plt.scatter(recommendations[~is_valid], np.ones_like(recommendations[~is_valid]), color='orange', s=60,
            marker='*', label='Recommended (by Engine)', zorder=10)

# Arrows from invalid input to recommendation
for x, rec in zip(inputs[~is_valid], recommendations[~is_valid]):
    plt.annotate('', xy=(rec, 1), xytext=(x, 0),
                 arrowprops=dict(facecolor='orange', shrink=0.05, width=2, headwidth=8, alpha=0.7))
    # Annotate a few for demonstration
    if np.isclose(x, 0.1) or np.isclose(x, 0.4):
        plt.text(x, 0.05, f"{x:.2f}→{rec:.2f}", fontsize=9, color='darkorange', rotation=50)

# Formatting
plt.xlabel('Input Configuration Value', fontsize=13)
plt.ylabel('Validity (1=Valid, 0=Invalid)', fontsize=13)
plt.title('Advanced Validation and Recommendation Visualization (with Engine)', fontsize=15)
plt.legend(loc='upper left')
plt.ylim(-0.15, 1.15)
plt.xlim(-0.05, 1.05)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
